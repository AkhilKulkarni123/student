# Major Commands Learned This Year:

ls - Lists files and folders in the current directory
cd - Changes directory (e.g., cd Documents or cd .. to go up one level)
pwd - Shows your current directory path
mkdir - Creates a new folder (e.g., mkdir project)
touch - Creates a new file (e.g., touch script.py)

File Operations

cp - Copies files (e.g., cp file1.txt file2.txt)
mv - Moves or renames files (e.g., mv old.txt new.txt)
rm - Deletes files (e.g., rm file.txt)
cat - Displays file contents in the terminal
nano or vim - Opens text editors in the terminal

Programming & Running Code

python or python3 - Runs Python files (e.g., python3 script.py)
java - Runs compiled Java programs
javac - Compiles Java files
node - Runs JavaScript files with Node.js
make - Runs build instructions from a Makefile (automates compilation)

Helpful Utilities

clear - Clears the terminal screen
man - Shows the manual/help for a command (e.g., man ls)
--help - Shows help for most commands (e.g., python --help)
history - Shows previous commands you've typed
grep - Searches for text within files

VS Code Specific

code . - Opens the current folder in VS Code
code filename - Opens a specific file in VS Code

# What I Want to Learn Next in Computer Science

Looking back at what we've covered in AP CSP, I feel like I've built a solid foundation in computational thinking and understanding how computing impacts society. But honestly, that's just scratching the surface of what's out there in CS, and I'm excited to dive deeper into several areas.
# Programming Languages
Right now, I'm comfortable with the basics we learned in class (JavaScript and Python), but I want to expand my language toolkit. Python is definitely at the top of my list if we didn't focus on it heavily – it's everywhere in data science, AI, and automation, which all sound really interesting. I'd also love to learn Java since it's used in AP CSA and is huge in enterprise software and Android development.
Down the line, I'm curious about lower-level languages like C++ or Rust to understand how memory management works and to write more performance-intensive programs. And SQL seems essential since basically every app needs to interact with databases somehow.
# Technical Skills
Web development is something I want to get better at. We touched on HTML/CSS, but I want to really understand frameworks like React or Vue and how to build full-stack applications with both frontend and backend components. Learning about APIs and how different services communicate would tie into this too.
Data structures and algorithms is another big one. AP CSP introduced lists and basic algorithms, but I know there's so much more – trees, graphs, hash tables, sorting algorithms, Big O notation. Understanding these deeply would help me write more efficient code and prepare for technical interviews.
I'm also really interested in cybersecurity. We talked about encryption and security concerns in CSP, but I want to learn about penetration testing, how to write secure code, and understand common vulnerabilities. It feels like such a critical skill in today's world.
# Soft Skills
On the softer side, collaboration and version control with Git/GitHub is something I want to master. We might have used it a bit, but really understanding branching, merge conflicts, and collaborative workflows seems crucial for working on any real team project.
Project management and documentation is another area for growth. Being able to break down large projects, estimate timelines, and write clear documentation that other people (or future me) can understand is something I've struggled with.
I also want to improve my debugging skills and develop better problem-solving strategies. Sometimes I get stuck and just randomly change things hoping something works – I need to be more systematic and learn to use debugging tools effectively.
# Specialized Topics
Looking ahead, I'm fascinated by machine learning and AI. The ethical implications we discussed in CSP made me curious about how these systems actually work and how to build them responsibly.
Mobile app development also appeals to me – either native iOS/Android development or cross-platform solutions. Almost everyone interacts with technology primarily through their phones, so understanding that ecosystem seems valuable.
Finally, I want to learn more about hardware and embedded systems. Maybe Arduino or Raspberry Pi projects? Understanding how software interfaces with physical hardware would give me a more complete picture of computing. (a little extra possibly)

Ultimately, CS is such a vast field that it's almost overwhelming, but that's also what makes it exciting. I think my next step is choosing one or two areas to focus on deeply rather than trying to learn everything at once, but I'm keeping an open mind about where my interests lead me.

# Main Achievments/Accomplishments and Learning Objectives Completed in AP CSP Trimester 1

Major Things Learned and Accomplishments
Reflecting on this year in AP CSP, I realize I've grown way beyond just learning syntax and basic programming concepts. I've developed real-world skills that professional developers actually use every day, and honestly, that feels pretty significant.
Repository Architecture and File Structure
One of the biggest breakthroughs for me was understanding repository architecture – like actually getting how a project is organized rather than just having a random collection of files. I learned that there's a logic to how folders are structured: you've got your source code in /src, configuration files at the root level, documentation in /docs, and maybe a /public or /assets folder for static resources, with things like west-coast.html or submodule1.md and questhome.md.
Understanding how different files work together was a game-changer. Before, I'd see files like package.json, .gitignore, or README.md and just... ignore them or treat them like mysterious config stuff. Now I understand that package.json manages dependencies and scripts, .gitignore tells Git what not to track, and configuration files like .env store environment variables and sensitive data that shouldn't be committed.
I also learned how files are connected and import each other – how a component file might import functions from a utility file, how CSS files link to HTML, how modules export and import between each other. Seeing that web of connections made everything click. It's not just isolated files; it's an ecosystem where everything has a purpose and relationship to other parts of the codebase.
Advanced Commands and Development Workflow
Moving beyond basic commands like git add and git commit, I learned more advanced Git operations that made me feel like I actually knew what I was doing. Commands like:

git branch and git checkout -b for creating and switching branches
git merge and dealing with merge conflicts (which was scary at first but satisfying once I got it)
git pull and git fetch to understand the difference between them
git log and git status to track what's happening in my repository
git reset and git revert for when things go wrong

I also learned how to quickly get localhost and server addresses running. Understanding that npm start or python -m http.server could spin up a local development server, and knowing where to find the localhost address (usually localhost:3000 or 127.0.0.1:8000) saved me so much time. I learned to check terminal output for the exact port, and how to access my site from other devices on the same network using my local IP address – that was cool for testing on mobile devices.
Styling and Customization
Learning how to change the theme of websites taught me a lot about CSS architecture and design systems. I learned how to:

Modify CSS variables or theme files to change color schemes globally
Override default styles by understanding CSS specificity and the cascade
Work with frameworks that have built-in theming (understanding where theme files live and how to customize them)
Use browser dev tools to inspect elements and test style changes in real-time before committing them to code

This wasn't just about making things look pretty – it taught me about maintainable code, because a well-structured theme system means you can change the entire look of a site by modifying just a few variables rather than hunting through hundreds of style rules.
Scrum Master Experience
Being Scrum Master was probably one of my most valuable experiences because it taught me skills that have nothing to do with coding but everything to do with software development. I learned how to:

Facilitate stand-up meetings and keep them focused and time-boxed
Maintain the sprint backlog and help the team prioritize tasks
Remove blockers – whether that meant helping teammates debug issues, clarifying requirements, or coordinating with other teams
Run sprint retrospectives where we reflected on what went well and what needed improvement
Balance being supportive versus being pushy when the team needed to stay on track

The soft skills I developed – communication, organization, conflict resolution, time management – were just as important as technical skills. I learned that building software is a team sport, and someone needs to make sure everyone's moving in the same direction. It also taught me empathy; when you're responsible for the team's success, you really pay attention to when people are struggling or overwhelmed.
Pull Requests and Code Review
Making PRs (Pull Requests) taught me about professional development workflows. I learned to:

Write clear, descriptive PR titles and descriptions explaining what changed and why
Reference related issues or tickets in my PRs
Request reviews from specific team members
Respond to code review feedback constructively (even when someone points out mistakes – which stung at first but made me a better developer)
Use the PR as a place for discussion about implementation decisions
Keep PRs focused and relatively small rather than dumping hundreds of lines of changed code

This process taught me that code isn't just about making something work – it's about making something that other people can understand, maintain, and build upon. Writing code that passes review means writing clean, documented, well-organized code.
How It All Comes Together
What's amazing is how all these pieces connect. Understanding repository architecture helps me make better PRs because I know where things belong. Advanced Git commands let me manage branches for different features. Server setup skills let me test my changes locally before pushing. And being Scrum Master taught me why all these technical practices exist – they're not bureaucracy; they're how teams build complex software without everything falling apart.
These aren't just "things I learned in a class." These are actual professional development practices, and having hands-on experience with them as a high school student feels like a huge head start. When I eventually interview for internships or jobs, I won't just say "I know how to code" – I can talk about real collaboration, version control workflows, project management, and team dynamics.
Looking back, I'm genuinely proud of how far I've come. I went from someone who could write basic programs to someone who understands how professional software development actually works. That's pretty cool.
