# Major Commands Learned This Year:

ls - Lists files and folders in the current directory
cd - Changes directory (e.g., cd Documents or cd .. to go up one level)
pwd - Shows your current directory path
mkdir - Creates a new folder (e.g., mkdir project)
touch - Creates a new file (e.g., touch script.py)

File Operations

cp - Copies files (e.g., cp file1.txt file2.txt)
mv - Moves or renames files (e.g., mv old.txt new.txt)
rm - Deletes files (e.g., rm file.txt)
cat - Displays file contents in the terminal
nano or vim - Opens text editors in the terminal

Programming & Running Code

python or python3 - Runs Python files (e.g., python3 script.py)
java - Runs compiled Java programs
javac - Compiles Java files
node - Runs JavaScript files with Node.js
make - Runs build instructions from a Makefile (automates compilation)

Helpful Utilities

clear - Clears the terminal screen
man - Shows the manual/help for a command (e.g., man ls)
--help - Shows help for most commands (e.g., python --help)
history - Shows previous commands you've typed
grep - Searches for text within files

VS Code Specific

code . - Opens the current folder in VS Code
code filename - Opens a specific file in VS Code

# Main Achievments/Accomplishments and Learning Objectives Completed in AP CSP Trimester 1

# Major Learnings and Accomplishments in AP CSP

This year in AP CSP, I've developed real-world skills that professional developers actually use, moving far beyond basic syntax.

## Repository Architecture and File Structure
I learned how projects are actually organized: `/src` for source code, `/docs` for documentation, `/public` for static resources, with config files at the root. Understanding files like `package.json` (dependencies and scripts), `.gitignore` (what Git shouldn't track), and `.env` (environment variables) was crucial. The biggest breakthrough was seeing how files connect—how components import utilities, how modules export and import between each other. It's an ecosystem where everything has purpose and relationships, not just isolated files.

## Advanced Git and Development Workflow
Beyond basic commands, I learned:
- `git branch`, `git checkout -b`, `git merge` (and handling merge conflicts)
- `git pull` vs `git fetch`, `git log`, `git status`
- `git reset` and `git revert` for fixing mistakes

I also learned to quickly spin up local servers with `npm start` or `python -m http.server`, find localhost addresses, and test on mobile devices using my local IP.

## Styling and Customization
Learning to change website themes taught me CSS architecture and design systems: modifying CSS variables for global changes, understanding specificity and the cascade, working with framework theming, and using browser dev tools to test changes. This taught me about maintainable code—good theme systems let you change an entire site's look by modifying just a few variables.

## Scrum Master Experience
Being Scrum Master taught me non-coding skills essential to software development: facilitating stand-ups, maintaining sprint backlogs, removing blockers, running retrospectives, and balancing support with keeping the team on track. I developed communication, organization, conflict resolution, and empathy—understanding that building software is a team sport.

## Pull Requests and Code Review
Making PRs taught me professional workflows: writing clear descriptions, referencing issues, responding constructively to feedback, and keeping PRs focused and small. This taught me that code isn't just about working—it's about being understandable, maintainable, and reviewable by others.

## How It All Connects
These skills interconnect: repository architecture helps me make better PRs, Git commands let me manage feature branches, server setup lets me test locally, and being Scrum Master showed me why these practices exist. These are actual professional development practices, and having hands-on experience as a high school student feels like a huge head start. When I interview for internships, I can discuss real collaboration, version control, project management, and team dynamics—not just "I know how to code."
